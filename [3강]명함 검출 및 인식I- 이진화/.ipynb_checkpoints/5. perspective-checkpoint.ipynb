### 영상의 기하학적 변환(geometric transformation)
- 영상을 구성하는 픽셀의 배치 구조를 변경함으로써 전체 영상의 모양을 바꾸는 작업
- Image registration, removal of geometric distortion, etc.

![](img/transform2.png)

##### 1. 찌그러진 명함 펴기

```python
import sys
import numpy as np
import cv2


src = cv2.imread('namecard1.jpg')

if src is None:
    print('Image load failed!')
    sys.exit()

w, h = 720, 400

# 모서리의 점을 잡아서 임의로 넣어준듯?
srcQuad = np.array([[325, 307], [760, 369], [718, 611], [231, 515]], np.float32)
dstQuad = np.array([[0, 0], [w-1, 0], [w-1, h-1], [0, h-1]], np.float32)

pers = cv2.getPerspectiveTransform(srcQuad, dstQuad)
dst = cv2.warpPerspective(src, pers, (w, h))

cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 1, figsize = (12, 12))

ax[0].imshow(src);    ax[0].set_title('before');     plt.axis('off'); ax[0].axis('off')
ax[1].imshow(dst);    ax[1].set_title('after perspective');     plt.axis('off'); ax[1].axis('off')
```
